```python
import pandas as pd
import os
```

```python
data_dir='csv_files'
data_dir
```

```python
os.listdir(data_dir)
```

```python
all_df=[]
files=os.listdir(data_dir)
for filename in files:
    full_path=os.path.join(data_dir,filename)
    df=pd.read_csv(full_path)
    all_df.append(df)
combined_df=pd.concat(all_df)
```

```python
combined_df.isnull().sum()
```

```python
combined_df['started_at']=pd.to_datetime(combined_df['started_at'])
combined_df['ended_at']=pd.to_datetime(combined_df['ended_at'])
combined_df.info()
```

```python
combined_df['ride_length']=(combined_df['ended_at']-combined_df['started_at']).dt.total_seconds()/60
```

```python
combined_df[['started_at','ended_at','ride_length']].head()
```

```python
combined_df['ride_length'].describe()
```

```python
clean_df=combined_df[combined_df['ride_length']>=1]
clean_df['ride_length'].describe()
```

```python
clean_df['day_of_week']=clean_df['started_at'].dt.day_name()
```

```python

clean_df = combined_df[combined_df['ride_length'] >= 1].copy()

clean_df['day_of_week'] = clean_df['started_at'].dt.day_name()

clean_df[['started_at', 'ride_length', 'day_of_week']].head()
```

```python
order=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
clean_df['day_of_week']=pd.Categorical(clean_df['day_of_week'],categories=order, ordered=True)
```

```python
clean_df.groupby(['member_casual','day_of_week'])['ride_length'].agg(['mean'],['count'])
```

```python
clean_df.groupby(['member_casual','day_of_week'], observed=False)['ride_length'].agg(['mean','count']).reset_index().sort_values(by='count', ascending=False)
```

```python
# 1. Verify the minimum ride length in our clean dataset
print(f"Minimum Ride Length: {clean_df['ride_length'].min()} minute")

# 2. Find the overall grand total count of rides for each user type
print("\n--- Grand Total Rides by User Type ---")
print(clean_df['member_casual'].value_counts())
```

```python
final_summary=clean_df.groupby(['member_casual','day_of_week'],observed=False)['ride_length'].agg(['mean','count']).reset_index().sort_values(by='count',ascending=False)
final_summary.to_csv('final_summary.csv',index=False)
```

```python
bike_summary=clean_df.groupby(['member_casual','rideable_type']).size().reset_index(name='total_trips')
bike_summary.to_csv('bike_summary.csv',index=False)
```

```python
#time summaries
clean_df['month']=clean_df['started_at'].dt.month_name()
clean_df['hour_of_day']=clean_df['started_at'].dt.hour
time_summary=clean_df.groupby(['member_casual','month','day_of_week','hour_of_day'],observed=False)['ride_length'].agg(['count','mean']).reset_index()
time_summary.columns=['member_casual', 'month', 'day_of_week', 'hour_of_day', 'total_trips', 'average_duration'] 
time_summary.to_csv('time-summary.csv', index=False)
```

```python
# 1. Group by user type and starting station, pulling the coordinates and ride lengths
station_summary = clean_df.groupby(['member_casual', 'start_station_name'], observed=False).agg(
    start_lat=('start_lat', 'mean'),
    start_lng=('start_lng', 'mean'),
    total_trips=('ride_length', 'count'),
    average_duration=('ride_length', 'mean')
).reset_index()

# 2. Drop any rows where the station name might be missing/blank
station_summary = station_summary.dropna(subset=['start_station_name'])

# 3. Export the final map summary to a CSV file
station_summary.to_csv('station_summary.csv', index=False)
```
